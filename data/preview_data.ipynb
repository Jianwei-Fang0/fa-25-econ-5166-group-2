# 📊 NTU YouBike 資料摘要

## "這個筆記本展示 NTU 區域 YouBike 資料的基本資訊和預覽。"

```python
import pandas as pd
import glob 
import os
```

## 📁 原始資料預覽

```python
# 讀取所有原始資料預覽
preview_files = glob.glob("..\\data\\raw/previews/*.csv")
preview_files.sort()

for file in preview_files:
    month = file.split('_')[0].split('/')[-1]
    print(f"\n📅 {month} 原始資料:")
    df = pd.read_csv(file)
    print(f"形狀: {df.shape}")
    display(df.head())
```

## 🚴‍♂️ NTU 處理後資料預覽

```python
ntu_files = glob.glob("..\\data/processed/ntu_youbike_data/*.csv")
ntu_files.sort()

for file in ntu_files:
    month = file.split('_')[-1].replace('.csv', '')
    print(f"\n📅 {month} NTU 資料:")
    df = pd.read_csv(file)
    print(f"形狀: {df.shape}")
    display(df.head())
```

## 🌊 站點流量資料預覽

```python
# 讀取站點流量資料
flow_files = glob.glob("..\\data/processed/station_flow/*.csv")
flow_files.sort()

for file in flow_files:
    month = file.split('_')[-1].replace('.csv', '')
    print(f"\n📅 {month} 站點流量:")
    df = pd.read_csv(file)
    print(f"形狀: {df.shape}")
    display(df.head())
```
