```python
import pandas as pd
df = pd.read_csv("weather_2024.csv")
df
```

```python
df.info()
```

```python
def take_date(str):
    parts = str.split("-")
    raw = "-".join(parts[1:])
    date = raw.split(".")[0]
    return date

df['日期'] = df['source_file'].apply(take_date)
df.head()
```

```python
df = df.drop(columns='source_file')
df.head()
```
